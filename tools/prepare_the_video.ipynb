```python
import yt_dlp

video_url = 'https://www.youtube.com/watch?v=m8Oyui5sGiI'
start_time = 50 * 60
end_time = 52 * 60
resolution = "480"
output_filename = "youtube_clip_50_52"

ydl_opts = {
    # Improved Format Selection:
    # This allows for more flexibility if a specific mp4 format is missing at 720p
    'format': f'bestvideo[height<={resolution}]+bestaudio/best[height<={resolution}]/best',
    'merge_output_format': 'mp4',
    'outtmpl': f'{output_filename}.%(ext)s',
    
    # Using firefox cookie for Auth
    'cookiesfrombrowser': ('firefox',),
    
    # Time Range Logic (The "Download Portion" part)
    'download_ranges': lambda info_dict, ydl: [{
        'start_time': start_time,
        'end_time': end_time,
    }],
    'force_keyframes_at_cuts': True,
    
    # FIX FOR "TLS Error / Code 187":
    # Using the 'ffmpeg' external downloader for the input stream specifically
    'external_downloader': {'default': 'ffmpeg'},
    'external_downloader_args': {
        'ffmpeg_i': [
            '-reconnect', '1', 
            '-reconnect_streamed', '1', 
            '-reconnect_delay_max', '5'
        ]
    },
    'quiet': False,
    'no_warnings': False,
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])
```
